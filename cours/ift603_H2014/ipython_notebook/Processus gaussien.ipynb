```python
%pylab inline
from numpy.random import *
from numpy.linalg import *
from numpy import *
import matplotlib.pyplot as plot

# Génère les points où on souhaite générer la valeur de la fonction
delta = 0.01
x = arange(-1.,1.,delta)

# Calcul de la matrice de Gram entre ces points, avec un noyau gaussien
def noyau_gaussien(x,sigma2):
    K = np.exp(-0.5*(x.reshape((-1,1)) - x.reshape((1,-1)))**2/sigma2)
    return K


# Génération de 10 fonctions avec sigma2 = 0.1
sigma2 = 0.1
K = noyau_gaussien(x,sigma2)
for i in range(10):
    y = random.multivariate_normal(np.zeros(len(x)),K)
    plot.plot(x,y)
```

```python
# Génération de 10 fonctions avec sigma2 = 1
sigma2 = 1
K = noyau_gaussien(x,sigma2)
for i in range(10):
    y = random.multivariate_normal(np.zeros(len(x)),K)
    plot.plot(x,y)
```

```python
# Génération de 10 fonctions avec sigma2 = 0.01
sigma2 = 0.01
K = noyau_gaussien(x,sigma2)
for i in range(10):
    y = random.multivariate_normal(np.zeros(len(x)),K)
    plot.plot(x,y)
```

```python
# On revient à un noyau gaussien avec sigma2 = 0.1
sigma2 = 0.1
K = noyau_gaussien(x,sigma2)

# Hyper-paramètre du bruit epsilon dans les cibles t (1/beta est la variance du bruit)
beta = 20. 

# Si t est une version bruité de y(x) (i.e. t = y(x) + epsilon), alors la matrice de covariance de t est :
C = K + 1/beta * eye(len(x))

# Génération de 10 fonctions y(x) + epsilon, avec sigma2 = 0.1
for i in range(10):
    t = random.multivariate_normal(np.zeros(len(x)),C)
    plot.plot(x,t)
```
