```python
%pylab inline
from numpy.random import *
from numpy.linalg import *
from numpy import *
import matplotlib.pyplot as plot

# Génération des données d'entraînement
X_1 = 0.5*randn(10,2) + array([[5,1]]) # Gaussienne centrée en mu_1=[5,1]
t_1 = ones(10)
X_2 = 0.5*randn(10,2) + array([[2,3]]) # Gaussienne centrée en mu_2=[2,3]
t_2 = zeros(10)

# Fusionne toutes les données dans un seul ensemble d'enraînement
X = vstack([X_1,X_2])
t = hstack([t_1,t_2])
#Mélange dans un ordre aléatoire
p = permutation(20)     
X = X[p,:]
t = t[p]

def sigm(a):
    return 1./(1. + exp(-a))

# Pour visualiser le classifieur
delta = 0.025
ix = np.arange(1.0, 6.0, delta)
iy = np.arange(0.0, 4.0, delta)
iX, iY = np.meshgrid(ix, iy)

# Initialisation du classifieur de la régression logistique
# (c'est valeurs sont choisies pour rendre l'illustration plus intéressante...
#  normalement on initialiserait à 0.)
w = array([1.,2.])
w_0 = -5.
eta = 0.5 # Normalement, on utiliserait une valeur plus petite

# Descente de gradient pour la régression logistique
plot.subplots(figsize=(20,25))
for it in range(20):
    # Calcul p(C_1 | x)
    o = sigm(w_0 + dot(X,w))
    
    # Calcul les gradients de w et w_0
    grad_w = dot((o-t),X)
    grad_w_0 = sum(o-t)
    
    # Mise à jour des paramètres
    w -= eta/20. * grad_w
    w_0 -= eta/20. * grad_w_0
    
    # Visualisation
    plot.subplot(5,4,1+it)
    dummy = plot.scatter(X_1[:,0],X_1[:,1])
    dummy = plot.scatter(X_2[:,0],X_2[:,1],marker='+')

    contour_out = sigm(dot(hstack([iX.reshape((-1,1)),iY.reshape((-1,1))]),w)+w_0).reshape(iX.shape)
    dummy = plot.contour(iX,iY,contour_out,[0.5], colors='k')
    plt.clabel(dummy, inline=1, fontsize=10)
```

```python
# Descent de gradient STOCHASTIQUE

# Initialisation du classifieur de la régression logistique
# (c'est valeurs sont choisies pour rendre l'illustration plus intéressante...
#  normalement on initialiserait à 0.)
w = array([1.,2.])
w_0 = -5.
eta = 0.5 # Normalement, on utiliserait une valeur plus petite

# Descente de gradient stochastique pour la régression logistique
plot.subplots(figsize=(20,20))
for it in range(1):
    for n in range(X.shape[0]):
        # Calcul p(C_1 | x)
        o = sigm(w_0 + dot(X[n:n+1,:],w))
        
        # Calcul les gradients de w et w_0
        grad_w = dot((o-t[n:n+1]),X[n:n+1,:])
        grad_w_0 = sum(o-t[n:n+1])

        # Mise à jour des paramètres
        w -= eta * grad_w
        w_0 -= eta * grad_w_0
        
        # Visualisation
        plot.subplot(5,4,1+(n*(it+1)))
        dummy = plot.scatter(X_1[:,0],X_1[:,1])
        dummy = plot.scatter(X_2[:,0],X_2[:,1],marker='+')
        dummy = plot.scatter(X[n,0],X[n,1],c='r',s=50.)

        contour_out = sigm(dot(hstack([iX.reshape((-1,1)),iY.reshape((-1,1))]),w)+w_0).reshape(iX.shape)
        dummy = plot.contour(iX,iY,contour_out,[0.5], colors='k')
        plt.clabel(dummy, inline=1, fontsize=10)
```
