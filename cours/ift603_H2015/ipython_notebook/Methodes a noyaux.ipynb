```python
%pylab inline
from numpy.random import *
from numpy.linalg import *
from numpy import *
import matplotlib.pyplot as plot

##
# Régression logistique à noyau (entraîné par descente de gradient stochastique)
##

# Génération des données d'entraînement
X_1_1 = 0.5*randn(5,2) + array([[5,1]]) # Gaussienne centrée en mu_1_1=[5,1]
X_1_2 = 0.5*randn(5,2) + array([[0,4]]) # Gaussienne centrée en mu_1_1=[0,4]
X_1 = vstack([X_1_1,X_1_2])
t_1 = ones(10)
X_2 = 0.5*randn(10,2) + array([[2,3]]) # Gaussienne centrée en mu_2=[2,3]
t_2 = zeros(10)

# Fusionne toutes les données dans un seul ensemble d'enraînement
X = vstack([X_1,X_2])
t = hstack([t_1,t_2])
#Mélange dans un ordre aléatoire
p = permutation(20)     
X = X[p,:]
t = t[p]

# Hyper-paramètre du noyau gaussien
sigma_2 = 0.5 

# Noyau gaussien
def k(x,x_prime):
      return exp(-dot(x-x_prime,x-x_prime)/(2*sigma_2))

    
# Pour visualiser le classifieur
delta = 0.025
ix = np.arange(-2.0, 7.0, delta)
iy = np.arange(-1.0, 6.0, delta)
iX, iY = np.meshgrid(ix, iy)
X_vis = hstack([iX.reshape((-1,1)),iY.reshape((-1,1))])

def sigm(a):
    return 1./(1. + exp(-a))

# Initialisation du classifieur de la régression logistique
a = zeros((20,))
w_0 = 0.
eta = 0.5 # Normalement, on utiliserait une valeur plus petite

# Pré-calcul de la matrice de Gram
K = array([[k(x,x_prime) for x_prime in X] for x in X])

# Pré-calcul de la matrice comparant X_vis à X
K_vis = array([[k(x,x_prime) for x_prime in X_vis] for x in X])

# Descente de gradient stochastique
plot.subplots(figsize=(20,20))
for it in range(1):
    for n in range(X.shape[0]):
        # Calcul p(C_1 | x)
        o = sigm(w_0 + dot(K[n,:],a))
        
        # Calcul les gradients de w et w_0
        grad_a = o-t[n]
        grad_w_0 = sum(o-t[n:n+1])

        # Mise à jour des paramètres
        a[n] -= eta * grad_a
        w_0 -= eta * grad_w_0
        
        # Visualisation
        plot.subplot(5,4,1+(n*(it+1)))
        dummy = plot.scatter(X_1[:,0],X_1[:,1])
        dummy = plot.scatter(X_2[:,0],X_2[:,1],marker='+')
        dummy = plot.scatter(X[n,0],X[n,1],c='r',s=50.)

        contour_out = sigm(dot(K_vis.T,a)+w_0).reshape(iX.shape)
        dummy = plot.contour(iX,iY,contour_out,colors='k')
        #dummy = plot.contourf(iX,iY,contour_out,cmap=cm.Greys)
        plt.clabel(dummy, inline=1, fontsize=10)
```
