```python
%pylab inline
import matplotlib.pyplot as plot
x = random.randn(100000)   # Génère 100000 échantillons d'une Gaussienne N(x|0,1)
dummy = plot.hist(x,100)
print "moyenne=",mean(x)
print "variance=",var(x)
```

```python
y = 2*x - 1 # Transforme les échantillons linéairement
dummy = plot.hist(y,100)
print "moyenne=",mean(y)    # Devrait être autour de 2*E[x] -1 = -1
print "variance=",var(y)    # Devrait êter autour de 2**2 * var[x] = 4
```

```python
import matplotlib.pyplot as plot
mu = array([0,0])
sigma = eye(2)
x = random.multivariate_normal(mu,sigma,100000).T # Génère 100000 échantillons d'une Gaussienne N(x|0,I) en 2 dimensions
dummy = plot.hist2d(x[0,:],x[1,:],50,cmap = cm.Greys)
axis('equal')
print "moyenne=",x.mean(axis=1)
print "covariance=\n",cov(x)
```

```python
import matplotlib.pyplot as plot
mu = array([1,2])
sigma = array([[2,0],[0,1]])
y = random.multivariate_normal(mu,sigma,100000).T # Génère 100000 échantillons d'une Gaussienne N(x|[1,2]^T,[[2,0],[0,1]]) en 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print "moyenne=",y.mean(axis=1)
print "covariance=\n",cov(y)
```

```python
import matplotlib.pyplot as plot
mu = array([1,2])
sigma = array([[1.5,1],[1,1.5]])
y = random.multivariate_normal(mu,sigma,100000).T # Génère 100000 échantillons d'une Gaussienne N(x|[1,2]^T,[[1.5,1],[1,1.5]]) en 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print "moyenne=",y.mean(axis=1)
print "covariance=\n",cov(y)
```

```python
mu = array([2,-1])
A = array([[3,0],[0,0.5]])
y = dot(A,x) + mu.reshape((2,1))
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print "moyenne=",y.mean(axis=1)
print 'devrait être',mu    # E[A x + mu] = A E[x]+mu = A(0)+mu = mu
print "covariance=\n",cov(y)
print 'devrait être\n',dot(A,A.T)   # cov[A x + mu] = A cov[x] A^T = A I A^T = A A^T
```
